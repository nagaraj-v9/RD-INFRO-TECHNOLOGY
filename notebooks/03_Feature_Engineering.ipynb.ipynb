### Objective

The objective of this notebook is to prepare the cleaned dataset for machine learning by converting categorical variables into numerical values and scaling numerical features.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../outputs/cleaned_airline.csv")
```

```python
df.head()
```

```python
df.head()
```

## 1. Dataset Information

```python
df.info()
```

## 2. Feature Encoding

from sklearn.preprocessing import LabelEncoder

```python
encoder = LabelEncoder()
```

```python
categorical_columns = [
    'gender',
    'customer_type',
    'type_of_travel',
    'class',
    'satisfaction'
]
```

```python
for column in categorical_columns:
    df[column] = encoder.fit_transform(df[column])
```

```python
df.head()
```

## 3. Feature Scaling

```python
scaler = StandardScaler()
```

```python
numerical_columns = [
    'age',
    'flight_distance',
    'departure_delay_in_minutes',
    'arrival_delay_in_minutes'
]
```

```python
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df.head()
```

### Explanation

Feature scaling standardizes numerical variables so that all features contribute equally during machine learning model training.

## 4. Save Processed Dataset

```python
df.to_csv("../outputs/processed_airline.csv", index=False)

print("Processed dataset saved successfully!")
```
