```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

%matplotlib inline
```

```python
df = pd.read_csv("../outputs/cleaned_airline.csv")
```

```python
df.head()
```

## 1. Dataset Overview

```python
df.shape
```

```python
df.info()
```

### Observation

The dataset contains over one hundred thousand passenger records with cleaned and standardized features ready for analysis.

## 2. Passenger Satisfaction Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,x='satisfaction')

plt.title("Passenger Satisfaction Distribution")

plt.tight_layout()

plt.savefig("../images/passenger_satisfaction.png",dpi=300)

plt.show()
```

### Business Insight

The dataset contains both satisfied and dissatisfied passengers, providing a balanced foundation for customer satisfaction analysis.

## 3. Passenger Satisfaction by Travel Class

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='class',
    hue='satisfaction'
)

plt.title("Passenger Satisfaction by Travel Class")

plt.tight_layout()

plt.savefig("../images/travel_class.png",dpi=300)

plt.show()
```

### Business Insight

Business Class passengers exhibit the highest satisfaction levels, while Economy passengers report comparatively lower satisfaction.

## 4. Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='age',
    bins=30,
    kde=True
)

plt.title("Age Distribution")

plt.tight_layout()

plt.savefig("../images/age_distribution.png",dpi=300)

plt.show()
```

### Business Insight

Most passengers belong to the adult age group, indicating the airline primarily serves working-age travelers.

## 5. Flight Distance Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='flight_distance',
    bins=30,
    kde=True,
    color='green'
)

plt.title("Flight Distance Distribution")

plt.tight_layout()

plt.savefig("../images/flight_distance.png",dpi=300)

plt.show()
```

### Business Insight

The airline handles a larger number of short and medium-distance flights than long-haul flights.

## 6. Inflight Wi-Fi Ratings

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='inflight_wifi_service'
)

plt.title("Inflight Wi-Fi Ratings")

plt.tight_layout()

plt.savefig("../images/wifi_rating.png",dpi=300)

plt.show()
```

### Business Insight

Passenger ratings indicate opportunities to improve onboard internet services for better customer experience.

## 7. Seat Comfort Ratings

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='seat_comfort'
)

plt.title("Seat Comfort Ratings")

plt.tight_layout()

plt.savefig("../images/seat_comfort.png",dpi=300)

plt.show()
```

### Business Insight

Seat comfort is an important contributor to passenger satisfaction and varies across customer experiences.

## 8. Departure Delay vs Passenger Satisfaction

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='satisfaction',
    y='departure_delay_in_minutes'
)

plt.title("Departure Delay vs Passenger Satisfaction")

plt.tight_layout()

plt.savefig("../images/departure_delay.png",dpi=300)

plt.show()
```

### Business Insight

Passengers experiencing higher departure delays generally report lower satisfaction levels.

## 9. Correlation Heatmap

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("../images/correlation_heatmap.png",dpi=300)

plt.show()
```

### Business Insight

The correlation heatmap highlights relationships between numerical features, helping identify variables that may influence passenger satisfaction.

# Final Conclusion

This exploratory data analysis identified key factors influencing passenger satisfaction.

Major findings include:

- Business Class passengers are generally more satisfied.
- Departure delays negatively impact customer experience.
